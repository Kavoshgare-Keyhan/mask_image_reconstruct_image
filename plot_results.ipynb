```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
mask_percentages = np.arange(0.1, 1.1, 0.1)
mask_percentages = np.append(mask_percentages,[.85,.95])
mask_percentages = np.sort(mask_percentages)
```

```python
rand_mask_avg_err = pd.read_csv('recons_average_mse_per_precision_random_mask.csv',header=None)
rand_mask_avg_err[0] = mask_percentages*100
# rand_mask_avg_err = rand_mask_avg_err.iloc[:,2].to_list()
grad_mask_avg_err = pd.read_csv('recons_average_mse_per_precision_grad_mask.csv',header=None)
grad_mask_avg_err[0] = mask_percentages*100

add_mask_avg_err = pd.read_csv('recons_average_mse_additive_masking_reconstruction.csv', header=None)
add_mask_avg_err.sort_values(by=0, ascending=False, inplace=True)
add_mask_avg_err[0] = mask_percentages*100
add_mask_avg_err
```

```python
rand_mask_avg_err
```

```python
plt.plot(rand_mask_avg_err.iloc[:,0], rand_mask_avg_err.iloc[:,2], marker='o', linestyle='-', color='orange', label='Rndom Masking Reconstruction Error')
plt.plot(grad_mask_avg_err.iloc[:,0], grad_mask_avg_err.iloc[:,2], marker='o', linestyle='-', color='yellow', label='Gradual Masking Reconstruction Error')
plt.plot(add_mask_avg_err.iloc[:,0], add_mask_avg_err.iloc[:,2], marker='o', linestyle='-', color='blue', label='Additive Masking Reconstruction Error')
plt.xlabel('Mask Percentage')
plt.ylabel('Error Rate')
plt.title('Error vs Mask Percentage')
plt.legend()
plt.grid(True)
# plot_path = 'recons/random/reconstruction_error_vs_mask_percentage.png'
plt.savefig('reconstruction_error_vs_mask_percentage.png')
plt.close()
```

```python
rand_mask_avg_class_err = pd.read_csv('recons_average_classification_accuracy_random.csv',header=None) #'recons_average_cross_entropy_error_random_mask.csv'
grad_mask_avg_class_err = pd.read_csv('recons_average_classification_accuracy_gradual.csv',header=None) #'recons_average_cross_entropy_error_grad_mask.csv'
add_mask_avg_class_err = pd.read_csv('recons_average_classification_error_additive_masking.csv', header=None)
add_mask_avg_class_err.sort_values(by=0, ascending=False, inplace=True)
```
